```python
class Solution(object):
    def strStr(self, haystack, needle):
        """
        :type haystack: str
        :type needle: str
        :rtype: int
        """
        # 如果 needle 是空字串，返回 0
        if not needle:
            return 0
        
        # 取得 haystack 和 needle 的長度
        haystack_len = len(haystack)
        needle_len = len(needle)
        
        # 如果 needle 長度大於 haystack，則不可能包含 needle，直接返回 -1
        if needle_len > haystack_len:
            return -1
        
        # 遍歷 haystack，檢查每個子字串是否與 needle 匹配
        for i in range(haystack_len - needle_len + 1):
            # 檢查從位置 i 開始的子字串是否等於 needle
            if haystack[i:i+needle_len] == needle:
                return i
        
        # 如果遍歷結束後還沒有找到 needle，則返回 -1
        return -1
```
