```python
class Solution(object):
    def removeElement(self, nums, val):
        index=0
        for i in range(0,len(nums)):
            if nums[i]==val:
                nums[i]=None
            else:
                index+=1
        while None in nums:
            nums.remove(None)
        return index
nums1 = [1, 1, 2, 3, 4,4,5]
sol=Solution()
print(sol.removeElement(nums1,4))
print(nums1)
```

```python
def removeElement(nums, val):
    k = 0  # Initialize k to count elements not equal to val
    
    for i in range(len(nums)):
        if nums[i] != val:
            nums[k] = nums[i]  # Move non-val element to the front
            k += 1  # Increment k
    
    return k
nums1 = [1, 1, 2, 3, 4,4,5,4,4,6]
print(removeElement(nums1,4))
print(nums1)
```
