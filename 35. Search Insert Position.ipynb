```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        
        mid = left + (right - left) // 2  # 中間索引
        
        # 檢查中間元素是否就是目標
        if arr[mid] == target:
            return mid
        
        # 如果目標值較大，忽略左半部分
        elif arr[mid] < target:
            left = mid + 1
        
        # 如果目標值較小，忽略右半部分
        else:
            right = mid - 1
    
    # 如果目標值不在數組中，返回 -1
    return -1

# 測試範例
arr = [1, 3, 5, 7, 9, 11, 13]
target = 11

result = binary_search(arr, target)

if result != -1:
    print(f"元素 {target} 位於索引 {result}")
else:
    print("元素未找到")
```

```python
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left=0
        right=len(nums)-1
        
        while left<=right:
            mid=left+(right-left)//2
            if target==nums[mid]:
                return mid
            elif target<nums[mid]:
                right=mid-1
            else:
                left=mid+1
        return -1
arr = [1, 3, 5, 7, 9, 11, 13]
target = 11
sol=Solution()
result = sol.searchInsert(arr, target)

if result != -1:
    print(f"元素 {target} 位於索引 {result}")
else:
    print("元素未找到")

    
```
